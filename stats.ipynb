```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_json("./data.jsonl", lines=True)

df['timestamp'] = pd.to_datetime(df['timestamp'])
df = df.set_index('timestamp')

ax = df.plot(y=['done', 'need updating', 'missing'], kind='line', stacked=True, color=['green', 'orange', 'red'])

ax.fill_between(df.index, df['done'], color='green', alpha=0.3)
ax.fill_between(df.index, df['done'], df['done'] + df['need updating'], color='yellow', alpha=0.3)
ax.fill_between(df.index, df['done'] + df['need updating'], df['done'] + df['need updating'] + df['missing'], color='red', alpha=0.3)

plt.title('Document Translation Progress Summary')
plt.xlabel('Timestamp')
plt.ylabel('Count')

plt.show()
```
